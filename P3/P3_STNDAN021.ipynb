```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
import cartopy.feature as cfeature 
```


# Figure 1: Antarctic Continent and Southern Ocean
## Projection: South Polar Stereo
### Description: This figure displays the Antarctic continent and the Southern Ocean starting from 60Â°S. The South Polar Stereo projection is chosen because it minimises distortion near the poles, providing a more accurate representation of high-latitude areas.

```python
# 1. Antarctic Continent and Southern Ocean
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': ccrs.SouthPolarStereo()})
ax.set_extent([-180, 180, -90, -60], ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, color='white')
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.gridlines(draw_labels=True, linestyle='--')
ax.set_title("Antarctic Continent and the Southern Ocean (60Â°S and below)")
ax.text(0, -75, "Southern Ocean", transform=ccrs.PlateCarree(), fontsize=12, color='blue', fontweight='bold', ha='center', bbox=dict(facecolor='white', alpha=0.7, edgecolor='none'))
plt.show()
```

# Figure 2: South Atlantic with Key Locations
## Projection: PlateCarree
### Description: This map covers the South Atlantic region from 20Â°S to 50Â°S, highlighting key coastal locations: Walvis Bay, Cape Town, Rio de Janeiro, and Montevideo. The Mercator projection is used as it preserves angles and directions, making it suitable for oceanic navigation.

```python
# 2. South Atlantic with Key Locations
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent([-60, 20, -50, -20], crs=ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, color='lightgray')
ax.add_feature(cfeature.OCEAN, color='lightblue')
ax.add_feature(cfeature.COASTLINE, edgecolor='black')

# Adding key locations
cities = {"Walvis Bay": (14.5, -22.9), "Cape Town": (18.4, -33.9),
          "Rio de Janeiro": (-43.2, -22.9), "Montevideo": (-56.2, -34.9)}

for city, (lon, lat) in cities.items():
    ax.plot(lon, lat, marker="o", color="red", markersize=5, transform=ccrs.PlateCarree())
    ax.text(lon + 1, lat, city, fontsize=10, transform=ccrs.PlateCarree())

ax.gridlines(draw_labels=True)
plt.title("South Atlantic with Key Locations")
plt.show()
```

# 3. False Bay Coastline Resolution Comparison
## Projection: Mercator
### Description: This figure illustrates the coastline resolution comparison for False Bay, South Africa, using three different levels of detail from the GSHHG database:

### 110m resolution (coarse) â The coastline is highly generalized, showing only the most basic shape.
### 50m resolution (intermediate) â More coastal features become visible, offering better representation.
### 10m resolution (high) â The coastline is highly detailed, closely matching the actual geography.
### This visualization demonstrates how different levels of resolution affect the clarity and precision of coastal mapping.

### The Mercator projection was used for this figure, it's generally used for navigation and oceanographic studies because it preserves angles and directions.It provides an accurate local representation of False Bay, making it a good choice for comparing coastline resolutions. Since False Bay is relatively close to the equator, Mercator projection does not introduce severe distortions in this region. It allows for easy comparison of coastline features across different resolutions without shape distortion.


```python
# 3. False Bay Coastline Resolution Comparison
resolutions = ['110m', '50m', '10m']
fig, axes = plt.subplots(1, 3, figsize=(15, 5), subplot_kw={'projection': ccrs.Mercator()})

for ax, res in zip(axes, resolutions):
    ax.set_extent([18.3, 18.9, -34.4, -34.1], crs=ccrs.PlateCarree())
    #ax.add_feature(cfeature.LAND, color='lightgray')
    ax.add_feature(cfeature.COASTLINE.with_scale(res), edgecolor='black')
    ax.gridlines(draw_labels=True)
    ax.set_title(f"False Bay - {res} resolution")
plt.show()
```
